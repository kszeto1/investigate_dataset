After wrangling the data, I'm deciding to drop the following:
* imdb_id

* overview

First, let's drop the columns that we won't use.

```python
df.drop(['imdb_id', 'homepage', 'tagline', 'keywords', 'overview'], axis=1, inplace=True)
```
